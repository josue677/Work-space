```python
%pylab inline
import scipy.optimize as optimize
from astropy.stats import LombScargle
```

```python
def Sine(t, amp = 1.0, offset = 0.0, period = 1.0, phase = 0.0):
    y = np.sin( 2*np.pi/period*(t-phase) )
    return amp/2 * y + offset
```

```python
def add_noise(x, amp=0.1):
    return asarray([i + (amp*(random.random()-0.5)) for i in x])
```

```python
# this is actually where the light "curve" gets defined
# note parameters below, both for t (observations) and Sine curve
t = add_noise(arange(0,30,0.31))
mag = Sine(t, amp=0.85, period=1.25, phase=0.45)
err = zeros(len(t)) + 0.05
```

```python
# how often are we observing?
plot(t,mag, 'bo')
```

Tough data! Lets generate a good guess using Fourier!

```python
frequency, power = LombScargle(t, mag).autopower()
plot(frequency, power)
#axis([0,0.01, 0, 0.9])
period = 1./frequency[argmax(power)]
print "period =", period
```

We'll use optimize.curve_fit to fit the data to the function we defined. This will (hopefully) return an array of the fitted parameters, which we'll save them directly into convience variables. Most examples keep the individual parameters in an array, which makes it easy to pass to other functions later.

```python
guess = array([1,0,1.25,0]) # amp = 1.0, offset = 0.0, period = 1.0, phase = 0.0
fit = optimize.curve_fit(Sine, t, mag, guess)
(amp, offset, period, phase) = fit[0]
print "amp =", amp, "offset =", offset, "period =", period, "phase =", phase
```

```python
x = arange(0, 30, 0.1)
plot(t,mag, 'bo')
plot(x, Sine(x, amp, offset, period, phase), 'g-')
```

A folded phase plot is much clearer

```python
plot((t % period)/period, mag, 'bo')
x = arange(0, period, period/1000) # one thousand points ought to look pretty smooth
plot( x / period, Sine(x, amp, offset, period, phase), 'g-' )
```
